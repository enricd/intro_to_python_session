```python
with open("aoc_day01_input.txt", "r") as file:    
    calories_input = file.read()

calories_input
```

```python
calories_input_list = calories_input.split("\n")
calories_input_list
```

```python
elves_calories = []
elf = []

for food in calories_input_list:
    if food == "":
        elves_calories.append(sum(elf))
        elf = []
    else:
        elf.append(int(food))
        
```

```python
# Part 1 result

max(elves_calories)
```
